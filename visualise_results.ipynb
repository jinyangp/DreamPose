```python
import os
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

## Visualise input pose, key frame image and the output generated image

```python
key_frame_img = mpimg.imread(os.path.join(os.getcwd(), 'demo', 'sample', 'key_frame.png'))
```

```python
print(frame_50_densepose_np[:1,:,:].shape)
print(frame_50_densepose_np[1:,:,:].shape)
```

### Frame 50

```python
frame_50_densepose_np = np.load(os.path.join(os.getcwd(), 'demo', 'sample', 'poses', 'frame_50_densepose.npy'))
frame_50_predicted_img = mpimg.imread(os.path.join(os.getcwd(), 'demo', 'sample', 'results', 'pred_#50.png'))
fig, axs = plt.subplots(1, 4, figsize = (24, 12))

axs[0].set_title("Key Frame")
axs[0].imshow(key_frame_img)
axs[1].set_title("DensePose np array - Part Channel")
axs[1].imshow(frame_50_densepose_np[:1,:,:].transpose(1,2,0))
axs[2].set_title("DensePose np array - U-coordinate of UV mapping Channel")
axs[2].imshow(frame_50_densepose_np[1:,:,:].transpose(1,2,0))
axs[3].set_title("Frame 50 Predicted Image")
axs[3].imshow(frame_50_predicted_img)
```
